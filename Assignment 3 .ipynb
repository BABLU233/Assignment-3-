6.  Explain the concept of generators in Python and how they are defined.?

-  Generators in Python
Generators are a type of iterable in Python that allow you to generate a sequence of values on-the-fly, rather than storing them all in memory at once. They're useful for working with large datasets or creating infinite sequences.

Defining Generators
Generators are defined using functions with the yield keyword instead of return. When a generator is called, it returns an iterator object that can be used to retrieve the generated values.


```python
# for Example

def sqare_no(n):
    result=[]
    for i in n:
        result.append(i**2)
    return result
```

```python
sqare_no([1,2,3,4,5])
```

```python
def sqare_no(n):
    result=[]
    for i in n:
        yield (i**2)
```

```python
sqare_no([1,3,4,5])
```

```python
gen=sqare_no([1,3,4,5])
```

```python
next(gen)
```

```python
next(gen)
```

7. What are the advantages of using generators over regular functions?
-    Advantages of Generators over Regular Functions
Generators offer several advantages over regular functions:

1. Memory Efficiency: Generators use less memory because they only generate values on-the-fly, whereas regular functions need to store all values in memory at once.
2. Lazy Evaluation: Generators only evaluate the next value when asked, which can be more efficient than computing all values upfront.
3. Flexibility: Generators can be used to create complex sequences or iterate over data in a customized way.
4. Improved Performance: Generators can be faster than regular functions because they avoid the overhead of storing and returning large datasets.
5. Handling Infinite Sequences: Generators can handle infinite sequences, which is not possible with regular functions.

When to use generators:

1. Large datasets: Generators are ideal for handling large datasets that don't fit in memory.
2. Infinite sequences: Generators are perfect for creating infinite sequences, such as random number generators or simulations.
3. Customized iteration: Generators can be used to iterate over data in a customized way, such as filtering or transforming data.

```python
# for Example fibonacci


def fibonacci(n):
  a,b=0, 1
  for i in range(n):
    yield a
    a,b=b, a+b
```

```python
gen=fibonacci(10)
```

```python
next(gen)
```

```python
next(gen)
```

```python
next(gen)
```

```python
next(gen)
```


8.   What is a lambda function in Python and when is it typically used?

-  Lambda Functions in Python
A lambda function is a small, anonymous function that can take any number of arguments, but can only have one expression. It's defined using the lambda keyword and is typically used for short, one-time use functions.   
When to use lambda functions:

1. One-time use: Lambda functions are perfect for one-time use cases, such as event handling or data processing.
2. Simple operations: Lambda functions are suitable for simple operations that don't require complex logic.
3. Higher-order functions: Lambda functions can be passed as arguments to higher-order functions, such as map(), filter(), or reduce().   Common use cases:

1. Data processing: Lambda functions can be used to process data in a concise way, such as sorting or filtering data.
2. Event handling: Lambda functions can be used as event handlers, such as button clicks or keyboard input.
3. Functional programming: Lambda functions can be used to create functional programming constructs, such as mapping or reducing data.


```python
#for Example lambda

add=lambda x,y:x+y
print(add(5,5))


```

```python
# using map() in lambda

num=[1,2,3,4,5,6]
square = list(map(lambda x:x**2,num))
print(square)
```

9. Explain the purpose and usage of the `map()` function in Python?

-   map() Function in Python
The map() function applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a map object, which is an iterator.

Syntax:

map(function, iterable)


Purpose:

1. Apply a function: map() applies a given function to each item of an iterable.
2. Transform data: map() can be used to transform data by applying a function to each item.

Usage:

1. Simple transformation: map() can be used to perform simple transformations, such as squaring numbers or converting strings to uppercase.
2. Data processing: map() can be used to process data in a concise way, such as converting data types or applying a calculation.

```python
# for example


num1=[1,2,3,4,5,6]
num2=[7,8,9,10,11,12]
multiplication= list(map(lambda x,y:x*y,num1,num2))
print(multiplication)

```

map() Function

```python
# Q7 Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
file_path = 'example.txt'
for line in read_file_lines(file_path):
    print(line)
```

```python
#Q9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32


celsisus = [0, 10, 20, 30, 40, 50]
fahrenheit = list(map(celsius_to_fahrenheit, celsisus))
print(fahrenheit)
```

```python
#Q10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    vowels = 'aeiouAEIOU'
    return filter(lambda x: x not in vowels, string)

string = "Hello, World!"
filtered_string = ''.join( remove_vowels(string))
print(filtered_string)
```

```python
# Q11 Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

#order no       book title and author           quantity    price per item
#34587         learning python, mark lutz          4               40.95
#98762         head first python ,paul berry       3               24.95
#77226         Einfurhrung python3, bernd klein    3               32.95
#88112         programming python ,mark lutz       5               56.80




#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.


order_list = [(34587, 40.95, 4), (98762, 24.95, 3), (77226, 32.95, 3), (88112, 56.80, 5)]

def calculate_order_totals(order_list):
    result = []
    for order_no, price_per_item, quantity in order_list:
        product = price_per_item * quantity
        if product < 100:
            product += 10
        result.append((order_no, product))
    return result

# Calculate and print the order totals
result = calculate_order_totals(order_list)
print(result)
```
